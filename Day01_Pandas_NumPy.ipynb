# **`NumPy Practice`**

```python
import numpy as np
```

```python
names = np.array(["Amit", "Riya", "John", "Sara", "Dev"])
ages = np.array([22, 25, 24, 23, 26])
scores = np.array([88, 92, 79, 95, 85])
cities = np.array(["Delhi", "Mumbai", "Delhi", "Bangalore", "Mumbai"])

print("\n=== Original NumPy Dataset ===")
print("Names :", names)
print("Ages  :", ages)
print("Scores:", scores)
print("Cities:", cities)
```

```python
print("\n=== Basic Operations ===")
print("Ages + 2:", ages + 2)
print("Scores * 2:", scores * 2)
print("Scores / 10:", scores / 10)

# Slicing
print("\nAges first 3:", ages[:3])
print("Scores last 2:", scores[-2:])
```

```python
print("\n=== NumPy Statistics ===")
print("Average Age:", np.mean(ages))
print("Max Score:", np.max(scores))
print("Min Score:", np.min(scores))
print("Std Deviation of Score:", np.std(scores))
print("Median Score:", np.median(scores))
```

```python
print("\n=== Filtering ===")

# Score > 85
high_scores = scores > 85
print("Scores > 85:", scores[high_scores])
print("Names with Score > 85:", names[high_scores])

# City = Delhi
delhi_students = cities == "Delhi"
print("\nDelhi Students:", names[delhi_students])

# Combined Condition: Age > 23 AND Score > 85
cond = (ages > 23) & (scores > 85)
print("\nAge > 23 & Score > 85:", names[cond])
```

```python
# Create matrix of Age & Score
matrix = np.vstack((ages, scores))
print("\n=== Matrix ===")
print(matrix)

print("\nTranspose:")
print(matrix.T)

print("\nDot Product of Age row and Score row:")
print(np.dot(ages, scores))
```

```python
scores_with_nan = scores.astype(float)
scores_with_nan[2] = np.nan  # inserting missing value

print("\n=== Scores with NaN ===")
print(scores_with_nan)

# Replace NaN with mean
mean_val = np.nanmean(scores_with_nan)
filled_scores = np.nan_to_num(scores_with_nan, nan=mean_val)

print("NaN Replaced with Mean:", filled_scores)
```
